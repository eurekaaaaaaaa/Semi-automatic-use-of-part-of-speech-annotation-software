```python
from pynput.keyboard import Key, Controller
from pynput import keyboard
from ctypes import *
import pyautogui
import time
import pyperclip
```

```python
#跑就行了的函数
def on_press2(key):
    
    try:
        if key.char in ['1', '2', '3', '4', '5', '6']:
            # 根据按键执行相应操作
            if key.char in ['1', '2', '3']:
                pyautogui.press('backspace')
                time.sleep(0.5)
                pyautogui.click(1000, 750, clicks=1, interval=0.0, button='left')
                value = {'1': '0.75', '2': '0.5', '3': '0.25'}[key.char]
                pyautogui.write(value)
            else:
                pyautogui.press('backspace')
                time.sleep(0.5)
                pyautogui.click(1000, 700, clicks=1, interval=0.0, button='left')
                value = {'6': '0.75', '5': '0.5', '4': '0.25'}[key.char]
                pyautogui.write(value)
            return False  # 停止监听
    except AttributeError:
        pass

def wait_for_pos_neg():
    # 设置监听器
    with keyboard.Listener(on_press=on_press2) as listener:
        listener.join()  # 阻塞直到监听器停止
```

```python
#跑就行了的函数*2
def on_press(key):
    if key == keyboard.Key.enter:
        # 当按下回车键时，停止监听
        return False

def wait_for_enter():
    # 设置键盘监听器
    with keyboard.Listener(on_press=on_press) as listener:
        listener.join()  # 阻塞，直到监听器停止
    # 监听器停止后，程序会在这里继续执行
#     print("回车键已被按下，程序继续执行。")
```

```python
#跑就行了的函数*3

def get_clipboard_text():
    pyautogui.hotkey('ctrl', 'c')  # 复制选中的内容到剪贴板
    time.sleep(0.5)  # 等待剪贴板内容更新
#     print(pyperclip.paste())
#     print(pyperclip.paste().startswith("P"))
    return pyperclip.paste()  # 返回剪贴板的内容

def click_and_get_text(x, y):
    pyautogui.doubleClick(x, y)  # 点击特定位置
    return get_clipboard_text()
        

def cixing():
    # 获取两个文本框的内容
    a = click_and_get_text(960, 495)
    
    # 判断条件并触发相应的函数
    
    if a =="adj":
        value = "a"
    elif a == "noun":
        value = "n"
    elif a == "verb":
        value = "v"
    else:
        pyautogui.click(1000, 550)
        wait_for_enter()
        return
    
    #点击和输出文本内容
    pyautogui.click(1000, 550) 
    pyautogui.write(value)
```

```python
#跑就行了的函数*3

#正负倾向判断
def emotion():
    # 获取两个文本框的内容
    a = click_and_get_text(960, 800)
    b = click_and_get_text(960, 845)
    
    # 判断条件并触发相应的函数
    try:
        b_int = int(b)
        if a.startswith('N'):
            pyautogui.click(1000, 750) 
            if 1 <= b_int <= 3:
                value = "3"
            elif 4 <= b_int <= 6:
                value = "2"
            elif 7 <= b_int <= 10:
                value = "1"
            text_to_type = {'1': '0.75', '2': '0.5', '3': '0.25'}[value]
        if a.startswith('P'):
            pyautogui.click(1000, 700) 
            if 1 <= b_int <= 3:
                value = "4"
            elif 4 <= b_int <= 6:
                value = "5"
            elif 7 <= b_int <= 10:
                value = "6"
            text_to_type = {'4': '0.25', '5': '0.5', '6': '0.75'}[value]
            
    except ValueError:
        print("B is not a number or out of range")
    
    #输出文本内容
    pyautogui.write(text_to_type)

# # 执行函数
# time.sleep(2)
# trigger_function_based_on_text()
# # click_and_get_text(960, 800)
```

```python
#主函数

kb = Controller()
time.sleep(5)
i = 0
#标注窗口的任务列表的第二个词的方框对准（471，366），鼠标点击第一个词
height = 366
bias = 18.1
while i < 32 :
        
    #词性类别
    pyautogui.click(1000, 555, clicks=1, interval=0.0, button='left')
    #定义一个函数，在我等待我输入enter以后运行下一个代码块
    cixing()
    
    #是否存在多个词义
    pyautogui.click(1000, 600, clicks=1, interval=0.0, button='left')
    kb.press('0')
#     time.sleep(1)
    
    #正负面倾向
    emotion() 
    
    #保存
    pyautogui.click(1092, 356, clicks=1, interval=0.0, button='left')
    time.sleep(2)
   
    #确定
    
    pyautogui.click(973, 625, clicks=1, interval=0.0, button='left')
    time.sleep(1)

    #选择下一个词语
    pyautogui.click(471, height+i*bias, clicks=1, interval=0.0, button='left')
#     kb.press('n')
    time.sleep(1)
    

#    print("==================this is the : "+str(i)+" times click=================")
    i += 1
```
