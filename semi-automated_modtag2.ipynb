```python
from pynput.keyboard import Key, Controller
from pynput import keyboard
from ctypes import *
import pyautogui
import time
import pyperclip
import math
```

```python
#跑就行了的函数
def on_press2(key):
    
    try:
        if key.char in ['1', '2', '3', '4', '5', '6']:
            # 根据按键执行相应操作
            if key.char in ['1', '2', '3']:
                pyautogui.press('backspace')
                time.sleep(0.5)
                pyautogui.click(1000, 750, clicks=1, interval=0.0, button='left')
                time.sleep(1)
                value = {'1': '0.75', '2': '0.5', '3': '0.25'}[key.char]
                pyautogui.write(value)
            else:
                pyautogui.press('backspace')
                time.sleep(0.5)
                pyautogui.click(1000, 700, clicks=1, interval=0.0, button='left')
                value = {'6': '0.75', '5': '0.5', '4': '0.25'}[key.char]
                pyautogui.write(value)
            return False  # 停止监听
    except AttributeError:
        pass

def wait_for_pos_neg():
    # 设置监听器
    with keyboard.Listener(on_press=on_press2) as listener:
        listener.join()  # 阻塞直到监听器停止
```

```python
#跑就行了的函数*2
def on_press(key):
    if key == keyboard.Key.enter:
        # 当按下回车键时，停止监听
        return False

def wait_for_enter():
    # 设置键盘监听器
    with keyboard.Listener(on_press=on_press) as listener:
        listener.join()  # 阻塞，直到监听器停止
    # 监听器停止后，程序会在这里继续执行
#     print("回车键已被按下，程序继续执行。")
```

```python
#跑就行了的函数*3

def get_clipboard_text():
    pyautogui.hotkey('ctrl', 'c')  # 复制选中的内容到剪贴板
    time.sleep(0.5)  # 等待剪贴板内容更新
#     print(pyperclip.paste())
#     print(pyperclip.paste().startswith("P"))
    return pyperclip.paste()  # 返回剪贴板的内容

def click_and_get_text(x, y):
    pyautogui.doubleClick(x, y)  # 点击特定位置
    time.sleep(1)
    return get_clipboard_text()
        

def cixing():
    # 获取两个文本框的内容
    a = click_and_get_text(960, 495)
    
    # 判断条件并触发相应的函数
    
    if a =="adj":
        value = "a"
    elif a == "noun":
        value = "n"
    elif a == "verb":
        value = "v"
    else:
        pyautogui.click(1000, 550)
        wait_for_enter()
        return
    
    #点击和输出文本内容
    pyautogui.click(1000, 550) 
    pyautogui.write(value)
```

```python
#跑就行了的函数*3

#正负倾向判断
def emotion():
    # 获取两个文本框的内容
    a = click_and_get_text(960, 800)
    b = click_and_get_text(960, 845)
    
    # 判断条件并触发相应的函数
    try:
        b_int = int(b)
        if a.startswith('N'):
            pyautogui.click(1000, 750) 
            if 1 <= b_int <= 3:
                value = "3"
            elif 4 <= b_int <= 6:
                value = "2"
            elif 7 <= b_int <= 10:
                value = "1"
            text_to_type = {'1': '0.75', '2': '0.5', '3': '0.25'}[value]
        if a.startswith('P'):
            pyautogui.click(1000, 700) 
            if 1 <= b_int <= 3:
                value = "4"
            elif 4 <= b_int <= 6:
                value = "5"
            elif 7 <= b_int <= 10:
                value = "6"
            text_to_type = {'4': '0.25', '5': '0.5', '6': '0.75'}[value]
            
    except ValueError:
        print("B is not a number or out of range")
    
    #输出文本内容
    pyautogui.write(text_to_type)

# # 执行函数
# time.sleep(2)
# trigger_function_based_on_text()
# # click_and_get_text(960, 800)
```

```python
def auto_cixing_writer(cixing_list,i):
    pyautogui.doubleClick(1000, 542)
    value = cixing_list[i]
    pyautogui.write(value)
```

```python
def generate_cixing_list():
    # 初始化一个空列表
    cixing_list = []

    # 循环，直到列表长度达到32或用户输入不是1-8之间的数字
    while len(cixing_list) < 32:
        # 提示用户输入数字
        user_input = input("1（副词） 2（形容词） 3（动词） 4（副词）: ")

        # 检查用户输入是否是1-8之间的数字
        if user_input in ["1", "2", "3", "4", "5", "6", "7", "8"]:
            # 根据用户输入的数字，将相应的字符串添加到列表中
#             mapping = {"1": "n", "2": "a", "3": "d", "4": "v", "5": "ia", "6": "iv", "7": "in", "8": "id"}
            mapping = {"1": "lgn", "2": "lga", "3": "lgv", "4": "lgd", "5": "ia", "6": "iv", "7": "in", "8": "id"}
            cixing_list.append(mapping[user_input])
        else:
            # 如果用户输入不是1-8之间的数字，结束循环
            cixing_list.append(user_input)

    return cixing_list

# 调用函数并获取生成的列表
cixing_list = generate_cixing_list()

# 打印生成的列表
print(cixing_list)
```

```python
def auto_emo_type_writer(emo_type_list, i):
    pyautogui.doubleClick(1000, 790)
    value = emo_type_list[i]
    pyautogui.write(value)
```

```python
def generate_emo_type_list():
    # 初始化一个空列表
    emo_type_list = []

    # 循环，直到列表长度达到31或用户输入不是1-8之间的数字
    while len(emo_type_list) < 32:
        # 提示用户输入数字
        user_input = input("1（赞扬）2（责备）3（烦闷）4（悲伤）5（慌乱）6（愤怒）7（憎恶）: ")

        # 检查用户输入是否是1-8之间的数字
        if user_input in ["1", "2", "3", "4", "5", "6","7"]:
            # 根据用户输入的数字，将相应的字符串添加到列表中
            mapping = {"1": "PH", "2": "NN", "3": "NE", "4": "NB", "5": "NI", "6": "NA", "7": "ND"}
            emo_type_list.append(mapping[user_input])
        else:
            # 如果用户输入不是1-6之间的数字，将用户输入加入到列表中
            emo_type_list.append(user_input)

    return emo_type_list

# 调用函数并获取生成的列表
emo_type_list = generate_emo_type_list()

# 打印生成的列表
print(emo_type_list)
```
